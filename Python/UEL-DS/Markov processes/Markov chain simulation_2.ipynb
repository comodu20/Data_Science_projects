```python
# BACK to tutorial

#Simulation of Gambler's ruin

# In our case the gambler starts off with $50. To keep things simple, they can only bet in $1 dollar increments. They can only win or lose $1
# They will keep gambling until they either lose all their money (leave with $0) or if they win $100. 
# We can formalise this with the ff: Xg = gambling outcome = {$+1, (Prb(0.5); $-1, Prb(0.5))}
# This is a perfect example of a discrete stochastic process


#lets start with creating a func

def gamblers_ruin():
    gambling_money = 50
    gambling_goal = 100
    gambling_simulations = []
    
    while gambling_money in range(1, gambling_goal):
        bet_size = 1
        win_or_loss = random.randrange(-1, 2, step = 2)    #The randrange() method returns a randomly selected element from the specified range.
                                                           # 1st argument is the start (An integer specifying at which position to start.), 
                                                           # 2nd argument is the stop (An integer specifying at which position to end. this integer is not included in possible outcomes)
                                                           # 3rd argument is the steps (An integer specifying the incrementation. Default is 1; the step being 2 here allows the program to skip 0)
        gambling_money += bet_size * win_or_loss
        gambling_simulations.append(gambling_money)        #bcos we set the gambling_simulations = [] (an empty list at the start. So to add outcomes, we use the append func)
    return gambling_simulations
```

```python
plt.plot(gamblers_ruin())
plt.yticks(np.arange(-20, 120, 10))
plt.axhline(y=0, color = 'r', linestyle = '-')
plt.axhline(y=100, color = 'black', linestyle = '-')
plt.xlabel('Number of bets')
plt.ylabel('Winnings')
plt.title('Gambling Simulation');                            # when i exlude the semicolon at the end it outputs some text lines before the graph
                                                             # FROM GEMINI:  In some contexts (like Jupyter Notebooks), a semicolon at the end of a Matplotlib plotting command can suppress the display of extra output information (like the object returned by the plotting function).  It's not strictly necessary in most cases, but it's a stylistic choice some programmers make.
```

```python
# n-step transition matrix

def matrix_power (matrix, power):
    if power == 0:
        return np.identity (len(matrix))        # numpy.identity(n, dtype=None, *, like=None)[source] : Return the identity array. The identity array is a square array with ones on the main diagonal.
    elif power == 1:
        return matrix
    else:
        return np.dot(matrix, matrix_power(matrix, power-1)) # This matrix_power is part of the function we defined above
                                                            # library is numpy.linalg.matrix_power --> linalg.matrix_power(a, n): Raise a square matrix to the (integer) power n. see practice script (In [7] towards tthe end) for example
                                                            # The matrix_power function calculates the power of a matrix using a recursive approach. 
                                                            # return np.dot(matrix, matrix_power(matrix, power-1)):
                                                            # matrix_power(matrix, power-1): This is the recursive call. The function calls itself with the same matrix but with the power reduced by 1. This continues until one of the base cases (power 0 or 1) is reached.
```

```python
# Borrowing from Markov chain simulation_1 script. find stationary TPM

import scipy.linalg     # This line imports the linalg module from the scipy (Scientific Python) library.  scipy provides many useful functions for scientific computing, including linear algebra.
values, left = scipy.linalg.eig(mc.to_numpy(), right = False, left = True)  # This is the core of the code.  It calls the scipy.linalg.eig function with the ff arguments:
                                                                    # we know mc.to_numpy() is the input tpm. "right=False": This argument tells the function not to calculate the right eigenvectors. Right eigenvectors are often calculated, but in this case, the code explicitly sets right=False because it only needs the left eigenvectors.
                                                                    # left=True: This argument tells the function to calculate the left eigenvectors.
                                                                    # values: A 1D array containing the eigenvalues of mc.to_numpy().
                                                                    # left: A 2D array where the columns are the left eigenvectors of mc.to_numpy().
        
print("left eigen vectors = \n", left, "\n")  
print("eigen values = \n", values)
```

```python
# Borrowing from Markov chain simulation_1 script. find stationary TPM 2

# you can see that of the left eigen values only column one is equal to 1 which is our goal. therefore we use column 1 (indexed at  0) of the eigen vectors

pi = left[:,0]              #notation selects all rows and only first column (indexed at 0) in "left" (from "values, left = scipy.linalg.eig" above)
pi_normalised = [(x/np.sum(pi).real) for x in pi]     # np.sum(pi).real: This calculates the sum of the elements of pi and takes the real part. Because of floating point calculations, even if the components of pi should be real, there could be a tiny, imaginary component. Taking the real part is a way to address this.
                                                      # (x/np.sum(pi).real): Each element x of pi is divided by the sum of all elements in pi, effectively normalizing the vector so that its elements sum to 1.
                                                      # [...]: The list comprehension creates a new list pi_normalised containing the normalized elements.
print(pi_normalised, '\n', '\n', sum(pi_normalised))

#super closeand more accurate than In [116] output above
```
